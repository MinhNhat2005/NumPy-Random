```python
# lấy mẫu 2x3 từ phân phối logistic có giá trị trung bình là 1 và độ lệch chuẩn là 2
from numpy import random
x = random.logistic(loc=1, scale=2, size=(3,3))
print(x)
```

```python
#Hình ảnh phân phối logistic
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

data = random.logistic(size=1000)

sns.displot(data, kind="kde")
plt.show()
```

```python
# Sự khác biệt giữa phân phối logistic và phân phối chuẩn
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

data = {
    "normal": random.normal(scale=2, size= 1000),
"logistic": random.logistic( size= 1000)
}
sns.displot(data, kind="kde")
plt.show()
```
