```python
# Phân phối Chi bình phương
# Phân phối Chi bình phương được sử dụng làm cơ sở để xác minh giả thuyết.

# Nó có hai tham số:

# df- (mức độ tự do).

# size- Hình dạng của mảng trả về
```

```python
# Lấy mẫu cho phân phối chi bình thường với bậc tư do 2 và  kích thước 2x3
from numpy import random
x = random.chisquare(df=2, size=(2,3))
print(x)
```

```python
# Hình ảnh phân phối chi Square
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

data = random.chisquare(df= 1, size=1000)
sns.displot(data, kind="kde")
plt.show()
```
