```python
# Phân phối mũ
# Phân phối mũ được sử dụng để mô tả thời gian cho đến sự kiện tiếp theo, ví dụ như thất bại/thành công, v.v.

# Nó có hai tham số:

# scale- nghịch đảo của tỷ lệ (xem lam trong phân phối Poisson) mặc định là 1,0.

# size- Hình dạng của mảng trả về.
```

```python
# Lấy mãu phân phối mũ với tý lệ 2 và kích thước 2x3
from numpy import random
x = random.exponential(scale=2, size=(2,3))
print(x)
```

```python
# Hình dung phân phối mũ
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

data = random.exponential(size=1000)

sns.displot(data, kind="kde")
plt.show()
```

```python
# Mỗi quan hệ giữ Poisson và phân phối mũ 
# Phân phói Poisson liên quan đến số lần xảy ra của một sự kiện trong một khoảng thời gian trong khi phân phối mũ liên quan đến thời gian giữa các sự kiện này
```
