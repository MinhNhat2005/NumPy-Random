<span style="font-size:30px;">Phân phối chuẩn (Gaussian)</span>


```python
# Nó có ba tham số:

# loc- (Trung bình) nơi có đỉnh chuông.

# scale- (Độ lệch chuẩn) độ phẳng của phân phối đồ thị.

# size- Hình dạng của mảng trả về.
```

```python
# Tạo phân phối chuẩn ngẫu nhiên có kích thước 2x3
from numpy import random
x = random.normal(size=(2,3))
print(x)
```

```python
# Tạo 1 phân phối chuẩn ngẫu nhiên có kích thước 2x3 với trung bình là 1 và độ lệch chuẩn là 2
from numpy import random
x = random.normal(loc=1, scale=2, size=(2,3))
print(x)
```

```python
# Hình dung phân phối chuẩn
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

sns.displot(random.normal(size=1000), kind="kde")
plt.show()
```
