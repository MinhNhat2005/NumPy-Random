```python
# Tạo phân phối ngẫu nhiên 1x10 cho lần xuất hiện thứ 2
from numpy import random
x = random.poisson(lam = 2, size= 10)
print(x)
```

```python
# Hình dung phân phối Poisson 
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

sns.displot(random.poisson(lam = 2, size=1000))
plt.show()
```

```python
# sự khác biệt giữa phân phối chuẩn vào phân phối Poisson
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

data = {
    "normal": random.normal(loc= 50, scale= 7, size=1000),
    "poisson": random.poisson(lam= 50, size = 1000)
}
sns.displot(data, kind = "kde")
plt.show()
```

```python
# Sự khác nhau giữa phân phối nhị thức và poisson
from numpy import random 
import matplotlib.pyplot as plt 
import seaborn as sns

data = {
    "binomial": random.binomial(n = 1000, p = 0.01, size= 1000),
    "poisson": random.poisson(lam= 10, size= 1000)
}
sns.displot(data, kind="kde")
plt.show()
```
