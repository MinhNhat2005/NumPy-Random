```python
# Phân phối Rayleigh
# Phân phối Rayleigh được sử dụng trong xử lý tín hiệu.

# Nó có hai tham số:

# scale- (độ lệch chuẩn) quyết định độ phẳng của phân phối mặc định là 1.0).

# size- Hình dạng của mảng trả về.
```

```python
# vé 1 một phân phối Rayleigh có tỷ lệ 2 với kích thước 3x3
from numpy import random
x = random.rayleigh(scale=2, size=(3,3))
print(x)
```

```python
# hình dung phân phối Rayleigh
from numpy import random 
import matplotlib.pyplot as plt 
import seaborn as sns

data = random.rayleigh(size= 1000)
sns.displot(data, kind="kde")
plt.show()
```
