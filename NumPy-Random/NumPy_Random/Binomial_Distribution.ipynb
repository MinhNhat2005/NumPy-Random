<span style="font-size:30px;">Phân phối nhị thức</span>


```python
# cho 10 lần tung đồng xu tạo ra 10 điểm dữ liệu
import numpy as np
x = np.random.binomial(n = 10, p = 0.5, size= 10)
print(x)
```

```python
# hình dung phân phối nhị thức
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

sns.displot(random.binomial(n = 10, p = 0.5, size=  1000))
plt.show()
```

<span style="font-size:30px;">sự khác nhau giữa phân phối chuẩn và phân phối nhị thức</span>


```python
# phân phối chuẩn là liên tục trong khi phân phối nhị thức là rời rạc
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

data_normal = random.normal(loc=50, scale=5, size=1000)
data_binomial = random.binomial(n=100, p=0.5, size=1000)

sns.kdeplot(data_normal, label="Normal distribution")
sns.kdeplot(data_binomial, label="Binomial distribution")

plt.legend()
plt.show()
```
