<span style="font-size:30px;">Phân phối đồng đều</span>


```python
# Tạo mảng phân phối đều 2x3
from numpy import random
x = random.uniform(size=(2,3))
print(x)
```

```python
# Hình dung phân phối đồng đều
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns
data = random.uniform(size = (100))

sns.displot(data, kind= "kde")
plt.show()
```
